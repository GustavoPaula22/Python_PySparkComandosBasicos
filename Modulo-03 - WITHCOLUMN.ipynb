```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = (
    SparkSession.builder.appName('Treinamento-02')
    .config('spark.sql.repl.eagerEval.enabled', True)
    .getOrCreate()
)
```

```python
df = spark.read.parquet(r'C:\Meus\Projetos\Python\VsCode\Bases\DATASETS\LOGINS.parquet')
```

```python
df
```

```python
# este comando é utilizado para criar uma nova coluna no DF, pode ser um campo fixo ou um campo condicional
df.withColumn('pais', F.lit('Brasil'))
```

```python
# para utilizar a forma condicional
(
    df
    .withColumn('descricao_estado', F.when(df.estado == 'GO', 'Goias')
                                     .when(df.estado == 'MG', 'Minas Gerais')
                                     .when(df.estado == 'GO', 'Goias')
                                     .when(df.estado == 'SP', 'Sao Paulo')
                                     .otherwise('Demais'))
)
```
