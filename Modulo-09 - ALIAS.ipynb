```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
```

```python
spark = (
    SparkSession.builder.appName('Treinamento-02')
    .config('spark.sql.repl.eagerEval.enabled', True)
    .getOrCreate()
)
```

```python
autores = spark.read.parquet(r'C:\Meus\Projetos\Python\VsCode\Bases\DATASETS\AUTORES.parquet')
livros = spark.read.parquet(r'C:\Meus\Projetos\Python\VsCode\Bases\DATASETS\LIVROS.parquet')
clientes = spark.read.parquet(r'C:\Meus\Projetos\Python\VsCode\Bases\DATASETS\CLIENTES.parquet')
compras = spark.read.parquet(r'C:\Meus\Projetos\Python\VsCode\Bases\DATASETS\COMPRAS.parquet')
logins = spark.read.parquet(r'C:\Meus\Projetos\Python\VsCode\Bases\DATASETS\LOGINS.parquet')
```

```python
# podemos utilizar o comando ALIAS para poder renomear colunas novas ou colunas ja existentes
logins.groupBy(F.col('estado').alias('UF')).agg(F.count('*').alias('qtd'))
```

```python
# pode ser feito o alias no proprio DF para poder utilizar para fazer cruzamentos de varios DF
livros2 = spark.read.parquet(r'C:\Meus\Projetos\Python\VsCode\Bases\DATASETS\LIVROS.parquet').alias('livros')
compras2 = spark.read.parquet(r'C:\Meus\Projetos\Python\VsCode\Bases\DATASETS\COMPRAS.parquet').alias('compras')

compras2.join(other=livros2, on=F.col('compras.cd_livro') == F.col('livros.id'), how='inner')
```
